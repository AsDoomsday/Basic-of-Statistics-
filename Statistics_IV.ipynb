```python
import random
import pandas as pd
import matplotlib.pyplot as plt

"""
ðŸŽ² Probability Distribution of Rolling Two Dice

A probability distribution describes how probabilities are spread across 
the possible values of a random variable.

For discrete random variables, the distribution can be represented as a 
table showing each possible value along with its associated probability.

In this example, we simulate rolling two dice many times and compute the 
probability of each possible sum (from 2 to 12). The results are then 
visualized using a bar chart.
"""

# Number of trials (higher = more accurate probabilities)
num_trials = 100_000

# Initialize dictionary to store frequencies of sums (2 through 12)
sum_counts = {total: 0 for total in range(2, 13)}

# Run simulation
for _ in range(num_trials):
    die1 = random.randint(1, 6)   # Roll first die
    die2 = random.randint(1, 6)   # Roll second die
    total = die1 + die2           # Sum of dice
    sum_counts[total] += 1        # Record occurrence

# Convert results into a DataFrame
df = pd.DataFrame({
    "Sum": sum_counts.keys(),
    "Count": sum_counts.values()
})
df["Probability"] = df["Count"] / num_trials

# ðŸ“Š Visualization
plt.figure(figsize=(8, 5))
plt.bar(df["Sum"], df["Probability"], color="skyblue", edgecolor="black")
plt.title("Probability Distribution of Rolling Two Dice", fontsize=14, weight="bold")
plt.xlabel("Sum of Two Dice", fontsize=12)
plt.ylabel("Probability", fontsize=12)
plt.xticks(df["Sum"])
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

# Display probability table
print(df)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

"""
ðŸ“ˆ Probability Distribution for Continuous Variables (Probability Density)

Unlike discrete variables, continuous random variables do not have probabilities 
for exact values. Instead, we calculate probabilities over an interval or range 
of values. This is why we use a probability density function (PDF).

One of the most common continuous distributions is the Normal Distribution, 
also known as the Gaussian distribution. 

The probability density function of the Normal distribution is given by:

    f(x) = (1 / (Ïƒ * âˆš(2Ï€))) * e^(-(x - Î¼)Â² / (2ÏƒÂ²))

where:
    Î¼ = mean (center of the distribution)
    Ïƒ = standard deviation (spread of the distribution)

ðŸ‘‰ In this example:
- We first simulate 1000 samples drawn from a normal distribution with mean = 5.2 
  and standard deviation = 0.4, and plot their histogram.
- Then, we plot several theoretical normal distributions with different means 
  and standard deviations for comparison.
"""

# ðŸŽ² Generate random samples from a Normal distribution (mean = 5.2, std = 0.4)
x = np.random.normal(loc=5.2, scale=0.4, size=1000)

# Plot histogram of simulated data
plt.hist(x, bins=15, edgecolor='black')
plt.title("Histogram of Heights", fontsize=14, weight="bold")
plt.xlabel("Height (cm)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.show()

# ðŸ“Š Plot theoretical normal distributions
a = np.linspace(-10, 10, 1000)  # Range of values for PDF

density = norm.pdf(a)                   # Standard normal (mean=0, std=1)
density_mean3 = norm.pdf(a, loc=3)      # Normal with mean=3, std=1
density_std2 = norm.pdf(a, scale=2)     # Normal with mean=0, std=2

plt.figure(figsize=(8, 5))
plt.plot(a, density, label="Mean = 0, Std = 1")
plt.plot(a, density_mean3, label="Mean = 3, Std = 1")
plt.plot(a, density_std2, label="Mean = 0, Std = 2")
plt.title("Normal Distribution Curves", fontsize=14, weight="bold")
plt.xlabel("Variable", fontsize=12)
plt.ylabel("Density", fontsize=12)
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

"""
ðŸ“Š Poisson Distribution

The Poisson distribution is a discrete probability distribution that describes 
the probability of a given number of events occurring within a fixed interval 
of time or space.

It is characterized by a single parameter Î» (lambda), which represents the 
average number of events in that interval.

The probability mass function (PMF) of the Poisson distribution is:

    P(X = k) = (e^(-Î») * Î»^k) / k!

Where:
    - P(X = k) = probability of observing exactly k events
    - Î» = average number of events in the interval (mean rate)
    - e â‰ˆ 2.718 (Eulerâ€™s number)
    - k! = factorial of k

ðŸ‘‰ Example:
Suppose the average number of events is Î» = 5. We can use the Poisson distribution 
to calculate the probability of observing k events for k = 0, 1, 2, ..., 14.
"""

# Parameter (average rate of events)
Î» = 5  

# Range of possible event counts
x = np.arange(0, 15)

# Probability Mass Function (PMF)
pmf = poisson.pmf(x, Î»)

# ðŸ“Š Visualization
fig, ax = plt.subplots(figsize=(8, 5))
ax.stem(x, pmf, basefmt=" ")
ax.set_title("Poisson Distribution (Î» = 5)", fontsize=14, weight="bold")
ax.set_xlabel("Number of Events (k)", fontsize=12)
ax.set_ylabel("Probability", fontsize=12)
ax.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
